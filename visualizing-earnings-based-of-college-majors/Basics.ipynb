# Visualizing Earnings Based On College Majors

```python
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline

```

```python
recent_grads=pd.read_csv('recent-grads.csv')
```

```python
print(recent_grads.iloc[0])
```

```python
raw_data_count=recent_grads.shape
recent_grads=recent_grads.dropna()
cleaned_data_count=recent_grads.shape
print(raw_data_count)
print(cleaned_data_count)
```

Generating different scatter plots to explore the relations between different columns

```python
recent_grads.plot(x="Sample_size",y="Median",kind='scatter',title='Sample_size vs Median',figsize=(8,8))
```

```python
recent_grads.plot(x="Sample_size",y="Unemployment_rate",kind='scatter',title='Sample size vs Unemployment rate')
```

```python
recent_grads.plot(x="Full_time",y="Median",kind='scatter',title='Full time vs Median')
```

```python

recent_grads.plot(x="ShareWomen",y="Unemployment_rate",kind='scatter',title='ShareWomen vs Unemployment rate')
```

```python

recent_grads.plot(x="Men",y="Median",kind='scatter',title='Men vs Median')
```

```python

recent_grads.plot(x="Women",y="Median",kind='scatter',title='Women vs Median')
```

```python
recent_grads['Sample_size'].hist(bins=10,range=(0,3000))
```

```python
cols=['Sample_size','Median','Employed','Full_time','ShareWomen','Unemployment_rate','Men','Women']
fig=plt.figure(figsize=(5,12))
for i in range(1,8):
    ax=fig.add_subplot(8,1,i)
    ax=recent_grads[cols[i]].plot(kind='hist')
```

# Scatter  Matrix Plot

```python

pd.plotting.scatter_matrix(recent_grads[['Sample_size','Median']],figsize=(10,10))
pd.plotting.scatter_matrix(recent_grads[['Sample_size','Median','Unemployment_rate']],figsize=(10,10))
```
